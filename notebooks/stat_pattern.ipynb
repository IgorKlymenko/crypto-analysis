```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df.index, df['Price'],      label='Price',      alpha=0.6)
plt.plot(df.index, df['MA_7'],       label='MA 7-day',    linewidth=2)
plt.plot(df.index, df['MA_30'],      label='MA 30-day',   linewidth=2)
plt.plot(df.index, df['MA_90'],      label='MA 90-day',   linewidth=2)
plt.title("Price with 7-day, 30-day, 90-day Moving Averages for ETH")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np


plt.figure(figsize=(12,5))
plt.plot(df.index, df['Price'],      label='Price', alpha=0.5)
plt.plot(df.index, df['MA_7'],       label='7-day MA', linewidth=2)
plt.plot(df.index, df['MA_30'],      label='30-day MA', linewidth=2)
plt.plot(df.index, df['MA_90'],      label='90-day MA', linewidth=2)

plt.legend(); plt.title("Price & Moving Averages for 7, 30, 90 days, Lido"); plt.show()


y = np.log(df['Price'])
t = (df.index.map(pd.Timestamp.toordinal) - df.index.min().toordinal())/365
X = sm.add_constant(t)
model = sm.OLS(y.values, X).fit()
print("Daily log-price slope:", model.params[1])
plt.figure(figsize=(12,5))
plt.plot(df.index, y, alpha=0.4)
plt.plot(df.index, model.fittedvalues, 'r', linewidth=2, label='Trend')
plt.legend(); plt.show()
```

```python
import matplotlib.pyplot as plt

df_vol = df.copy().iloc[6:]

vol = df_vol['Vol_7']

mv = vol.rolling(window=90, min_periods=90).mean()
sv = vol.rolling(window=90, min_periods=90).std()


threshold = mv + 2*sv

plt.figure(figsize=(12,4))
plt.plot(df_vol.index, vol,        label='Vol_7', alpha=0.6)
plt.plot(df_vol.index, mv,         '--', color='green', label='90d Rolling Mean')
plt.plot(df_vol.index, threshold,  '--', color='red',   label='90d Mean + 2')
plt.title("7-Day Rolling Volatility with Dynamic Threshold, ETH")
plt.xlabel("Date")
plt.ylabel("Volatility (Std of Returns)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1-day raw market cap and its 30- & 90-day moving averages
mc = df['Market_cap']
ma30 = mc.rolling(30, min_periods=1).mean()
ma90 = mc.rolling(90, min_periods=1).mean()

plt.figure(figsize=(12,5))
plt.plot(df.index, mc,   label='Market_cap', alpha=0.5)
plt.plot(df.index, ma30, '--', label='30-day MA', linewidth=2)
plt.plot(df.index, ma90, '--', label='90-day MA', linewidth=2)
plt.title("Market Cap & Moving Averages")
plt.xlabel("Date"); plt.ylabel("Market_cap")
plt.legend(); plt.grid(True); plt.show()
```

```python

df = pd.read_csv("../preprocessed/ETH.csv", parse_dates=['Date'],
                date_format='%Y-%m-%d')

#df['Date'] = pd.to_datetime(df['Date'].str.replace(' UTC+0', ''))
df = df.set_index('Date').sort_index()


df2 = pd.read_csv("../preprocessed/Lido.csv", parse_dates=['Date'],
                date_format='%Y-%m-%d')

#df2['Date'] = pd.to_datetime(df2['Date'].str.replace(' UTC+0', ''))
df2 = df2.set_index('Date').sort_index()


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))

# Left plot - ETH
ax1.scatter(df['Price'], df['Market_cap'], alpha=0.4)
ax1.set_title("Market Cap vs. Price, ETH")
ax1.set_xlabel("Price")
ax1.set_ylabel("Market_cap")
ax1.grid(True)

# Right plot - df2
ax2.scatter(df2['Price'], df2['Market_cap'], alpha=0.4)
ax2.set_title("Market Cap vs. Price, Lido")
ax2.set_xlabel("Price")
ax2.set_ylabel("Market_cap") 
ax2.grid(True)

plt.tight_layout()
plt.show()

# Calculate correlations
corr1 = df['Price'].corr(df['Market_cap'])
corr2 = df2['Price'].corr(df2['Market_cap'])
print(f"ETH - Pearson corr(Price, Market_cap): {corr1:.3f}")
print(f"LDO - Pearson corr(Price, Market_cap): {corr2:.3f}")
```

```python
df = pd.read_csv("../preprocessed/ETH.csv", parse_dates=['Date'],
                date_format='%Y-%m-%d')

#df['Date'] = pd.to_datetime(df['Date'].str.replace(' UTC+0', ''))
df = df.set_index('Date').sort_index()


df2 = pd.read_csv("../preprocessed/Lido.csv", parse_dates=['Date'],
                date_format='%Y-%m-%d')

#df2['Date'] = pd.to_datetime(df2['Date'].str.replace(' UTC+0', ''))
df2 = df2.set_index('Date').sort_index()


# Calculate rolling correlations
roll_corr1 = df['Price'].rolling(30).corr(df['Market_cap'])
roll_corr2 = df2['Price'].rolling(30).corr(df2['Market_cap'])

# Plot both rolling correlations on same graph
plt.figure(figsize=(12,4))
plt.plot(roll_corr1, label='ETH')
plt.plot(roll_corr2, label='Lido')
plt.title("30-Day Rolling Correlation: Price vs Market Cap")
plt.xlabel("Date")
plt.ylabel("Correlation")
plt.legend()
plt.grid(True)
plt.show()

# Calculate overall correlations
corr1 = df['Price'].corr(df['Market_cap'])
corr2 = df2['Price'].corr(df2['Market_cap'])
print(f"ETH - Overall Pearson corr(Price, Market_cap): {corr1:.3f}")
print(f"LDO - Overall Pearson corr(Price, Market_cap): {corr2:.3f}")

# Print average rolling correlations
print(f"\nETH - Mean 30-day rolling correlation: {roll_corr1.mean():.3f}")
print(f"LDO - Mean 30-day rolling correlation: {roll_corr2.mean():.3f}")
```
